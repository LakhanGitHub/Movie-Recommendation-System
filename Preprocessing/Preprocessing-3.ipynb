```python
from tmdbv3api import Movie

tmdb_movie = Movie()

def get_genre(x): #pass the title of the movie
    genres = []
    result = tmdb_movie.search(x) #tile will serach
    movie_id = result[0].id #will mach id with title
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json() #data coveret into json format
    if data_json['genres']:
        genres_str = ' '
        for i in range(0, len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genres_str.join(genres)
    else:
        np.NaN #will retun nan
    
```

```python
import re

# Assuming your DataFrame is called `data`
# And the relevant column is named 'Cast and crew'

def extract_director(text):
    # This regex finds a name followed by '(director' — also works for '(director/screenplay)'
    match = re.search(r'([A-Za-z\s\.]+)\s*\(director', text)
    if match:
        return match.group(1).strip()
    return None

df_2018['director_name'] = df_2018['Cast and crew'].apply(extract_director)
```

# Actor 1

```python
#pd.set_option('display.max_colwidth', 200)

import pandas as pd
import re

def extract_actors(text):
    # Split into parts by semicolon
    parts = text.split(';')
    if len(parts) > 1:
        # All parts after the last semicolon likely contain actors
        actors_text = parts[-1]
        # Split by commas and strip spaces
        actors = [actor.strip() for actor in actors_text.split(',')]
        # Return the first three actors (or None if missing)
        return pd.Series(actors[:3] + [np.NaN] * (3 - len(actors)))
    else:
        return pd.Series([np.NaN, np.NaN, np.NaN])

# Apply to your DataFrame
df_2018[['actor_1_name', 'actor_2_name', 'actor_3_name']] = df_2018['Cast and crew'].apply(extract_actors)
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_American_films_of_2024'

df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
df
```

```python
df.dropna(subset='Title', inplace=True)
```

```python
df['Title'].isnull().sum()
```

```python
from tmdbv3api import Movie,TMDb

import json
import requests
tmdb = TMDb()
tmdb.api_key = '603e19ffe04402de9bf6b94cbf4ab870'


tmdb_movie = Movie()

def get_genre(title):
    try:
        result = tmdb_movie.search(title)
        if not result or not hasattr(result[0], 'id'):
            return np.nan

        movie_id = result[0].id

        response = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb.api_key}')
        data_json = response.json()

        if 'genres' in data_json:
            genres = [genre['name'] for genre in data_json['genres']]
            return ', '.join(genres)
        return np.nan

    except Exception as e:
        print(f"Error processing '{title}': {e}")
        return np.nan
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(x))
df
```

```python
df_2024 = df[['Title','Cast and crew', 'genres']]
```
